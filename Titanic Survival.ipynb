```python
#'Name','Sex','embark' etc are referred as features..
def bar_chart(feature): 
    survived = train[train['Survived']==1][feature].value_counts()
    dead = train[train['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
bar_chart('Sex')
```

```python
bar_chart('Pclass') 
####The above chart conveys that the passengers in 3rd class were dead by majority. Similarly, Passengers in 1st class survived than the rest.
```

```python
bar_chart('Embarked')
##The Chart confirms a person aboarded from S more likely dead The Chart confirms a person aboarded from C slightly more likely survived
```

```python
#To check whether the person Survived /dead is married, not married, men, women or childeren etc.. Let us
# extract the information by combining the train and test dataset together.
train_test_data = [train,test]

for dataset in train_test_data:    
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
#we are extracting the title like 'Mr', 'Mrs', 'Miss' etc from the Name column which is present in the 
#train_test_data dataset.
```

```python
train['Title'].value_counts()
```

```python
test['Title'].value_counts()
```

```python
title_mapping = {"Mr": 0, "Mrs": 1, "Miss": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
for dataset in train_test_data:
    dataset['Title'] = dataset['Title'].map(title_mapping)
```

```python
test.head()
```

```python
bar_chart('Title')

#Inference: 'Mr: 0' has the high chance to die than 'Mrs: 1' and 'Miss: 2'.
```

```python
#Similarly, we can assign 'Male' to 0 and 'Female' to 1, than to specify male and female explicitly in the 
# 'Sex' column everytime.

sex_mapping = {"male" : 0, "female": 1}

for dataset in train_test_data:
    dataset['Sex'] = dataset['Sex'].map(sex_mapping)
```

```python
bar_chart('Sex')

#Inference: The Female has higher chance to survive than the male.
```

```python

##Binning/Converting Numerical Age to Categorical Variable

##feature vector map: child: 0, young: 1, adult: 2, mid-age: 3, senior: 4
for dataset in train_test_data:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0,
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 26), 'Age'] = 1,
    dataset.loc[(dataset['Age'] > 26) & (dataset['Age'] <= 36), 'Age'] = 2,
    dataset.loc[(dataset['Age'] > 36) & (dataset['Age'] <= 62), 'Age'] = 3,
    dataset.loc[ dataset['Age'] > 62, 'Age'] = 4
```
